# Group-level analysis: Hero vs. Villain
On average, do heroes have higher stats than villains?

Which stat is most differentiated between heroes and villains?

```python
# 1
powers_df = hero_df[["Name"] + powers].copy()

#
powers_df.loc[:, 'Alignment'] = hero_df['Alignment_x']
# count of powers per char
powers_df.loc[:, 'Power_Count'] = powers_df[powers].sum(axis=1)
powers_per_alignment = powers_df.groupby('Alignment')['Power_Count'].agg(['mean', 'sum', 'count']).reset_index()
powers_per_alignment.columns = ['Alignment', 'Avg_Power_Count', 'Total_Power_Count', 'Character_Count']

# drop the '-' unsignificant alignment
powers_per_alignment.drop(inplace=True, index=0)
powers_per_alignment


# conclusion: neutral chars > villains > heros, is the order of how powerful each category is
```

```python
# 2: the stat that is least common between villains and heros
'''
Figuring it out:
I need a dataframe of powers on rows,
then 2 cols Bad_AVG vs Good_AVG which contain AVG power_count of that stat in each category
then one last col of Good_AVG / Bad_AVG, let's call it ratio.
Lastly, take the stat with min score


update: I figured it's better to compare using difference too
'''

most_diff_stat_df = powers_df.groupby('Alignment')[powers].mean()

# drop the '-' unsignificant alignment
most_diff_stat_df = most_diff_stat_df.reset_index().drop(index=0)

# la transposÃ©e
most_diff_stat_df = most_diff_stat_df.T

# fixing index, rather taking bad, good
most_diff_stat_df.columns = most_diff_stat_df.iloc[0]
most_diff_stat_df = most_diff_stat_df[1:]

# renaming just for convention
most_diff_stat_df = most_diff_stat_df.rename(columns={'bad': 'bad_avg', 'good': 'good_avg'})

# calc comparison scores
# division if no 0 division included
most_diff_stat_df['ratio'] = most_diff_stat_df.apply(lambda row: row['good_avg'] / row['bad_avg'] if row['bad_avg'] != 0 else None, axis=1)

# abs of diff
most_diff_stat_df['diff'] = (most_diff_stat_df['good_avg'] - most_diff_stat_df['bad_avg']).abs()

most_diff_stat_df
```

```python
most_diff_stat_df.sort_values(['ratio', 'diff'], ascending=True)
```

```python
least_ratio_stat = most_diff_stat_df['ratio'].idxmin()
least_difference_stat = most_diff_stat_df['diff'].idxmin()

# final answer
print(f"Power with least ratio: {least_ratio_stat}")
print(f"Power with least difference: {least_difference_stat}")
```

Identify Outliers:
Are there any characters with extreme stats (e.g., very high strength but very low speed)?


```python
from scipy.stats import zscore


stats_df = hero_df[["Name"] + stats]

z_scores = stats_df[stats].apply(zscore)

outliers = z_scores[(z_scores > 3) | (z_scores < -3)].dropna(how='all')
outlier_characters = stats_df.loc[outliers.dropna(how="all").index]

sns.scatterplot(data=stats_df, x="Strength", y="Speed", label="Normal")
sns.scatterplot(data=outlier_characters, x="Strength", y="Speed", color='red', label="Outliers")
plt.show()

# no outliers
```

**Predicting characters' alignments**

While predicting characters' alignments based on powers or stats might be completely meaningless due to randomness of these features, I will still be putting a machine learning model to work just to validate my hypothesis (expecting an overfit model, poor accuracy, just a model trained to noise and not enough valuable training data in each alignment to begin with)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
ml_hero_df = hero_df.copy()

ml_hero_df = ml_hero_df[ml_hero_df['Alignment_x'].isin(['good', 'bad'])]
ml_hero_df['Alignment_x'] = ml_hero_df['Alignment_x'].map({'good': 1, 'bad': 0})
ml_hero_df.value_counts('Alignment_x')
```

```python
# features : stats
features = ['Intelligence_x', 'Strength', 'Speed', 'Durability_x', 'Power', 'Combat']
X = ml_hero_df[features]

# target
y = ml_hero_df['Alignment_x']

# handling missing values
X.fillna(X.mean(), inplace=True)

# norm stats
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# predictions
y_pred = model.predict(X_test)

# evaluation
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
```

# Clustering the characters based on their stats

```python
from sklearn.preprocessing import StandardScaler

clustering_data = hero_df[stats]
clustering_data_filled = clustering_data.fillna(clustering_data.mean())

scaler = StandardScaler()
clustering_data_scaled = scaler.fit_transform(clustering_data_filled)

clustering_data_scaled
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
hero_df['Cluster'] = kmeans.fit_predict(clustering_data_scaled)


hero_df.value_counts('Cluster')
```

```python
# Viz clusters
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
reduced_data = pca.fit_transform(clustering_data_scaled)

plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=hero_df['Cluster'], cmap='viridis')
plt.title('Clusters of characters based on stats and powers')
plt.colorbar(label='')

plt.show()
```

```python
import numpy as np
from math import pi

characters = ['Doctor Strange', 'Scarlet Witch', 'Jean Grey', 'Magneto']
stats = ["Intelligence_x", "Strength", "Speed", "Durability_x", "Power", "Combat"]

stats_df = hero_df[hero_df['Name'].isin(characters)][stats]
stats_df.index = hero_df[hero_df['Name'].isin(characters)]['Name']

categories = list(stats_df.columns)
N = len(categories)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()

# Close the radar chart by appending the first value to the end for each character
angles += angles[:1]
stats_df = stats_df.apply(lambda row: row.tolist() + [row.tolist()[0]], axis=1)
stats_df = pd.DataFrame(stats_df.tolist(), index=stats_df.index)

# plot radar
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# lore accurate colors :D
colors = {
    'Doctor Strange': '#1E90FF',  # Mystic blue
    'Scarlet Witch': '#FF2400',  # Scarlet red
    'Jean Grey': '#FFA500',      # Fiery orange
    'Magneto': '#800080',        # Deep purple
}

# plot each character
for i, (character, values) in enumerate(stats_df.iterrows()):
    ax.plot(angles, values, label=character, linewidth=2, linestyle='solid', color=colors[character])
    ax.fill(angles, values, alpha=0.25, color=colors[character])






# Customize the chart
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=12)

# Add a title and legend
plt.title('Comparison of Character Stats', size=16, y=1.1)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1), title="Characters")

# Display the plot
plt.tight_layout()
plt.show()
```
